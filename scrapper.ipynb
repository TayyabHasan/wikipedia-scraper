# 5 POS tagging

```python
# pos tagger
tags = nltk.pos_tag(tokens_lower)
tags
```

```python
# to count total of each class
counts = nltk.FreqDist(tag for (word,tag) in tags )
counts
```

# 6 Lemmatize the most frequently occurring 10 adjectives
Adjective tag in POS = JJ
[Source](https://medium.com/@gianpaul.r/tokenization-and-parts-of-speech-pos-tagging-in-pythons-nltk-library-2d30f70af13b#:~:text=preposition/subordinating%20conjunction-,JJ%20adjective%20%E2%80%98big%E2%80%99,-JJR%20adjective%2C%20comparative)

```python
# finding adjective frequency distribution
adjective_fd = nltk.FreqDist(word for (word,tag) in tags if tag == 'JJ')
adjective_fd
```

```python
# Reading top Ten most common words in adjective frequency distribution 
lemma_input = [ word for( word, value ) in adjective_fd.most_common(10) if word ]
```

```python
wordNet = nltk.WordNetLemmatizer()
lemma_output = [wordNet.lemmatize(word) for word in lemma_input]

df =  pd.DataFrame(list(zip(lemma_input,lemma_output)))
df
```

# 7 Stem the least frequently occuring 10 verbs
VBG verb, gerund/present participle taking
[Source](https://medium.com/@gianpaul.r/tokenization-and-parts-of-speech-pos-tagging-in-pythons-nltk-library-2d30f70af13b#:~:text=VBG%20verb%2C%20gerund/present%20participle%20taking) 

```python
snowball = SnowballStemmer(language='english')
```

```python
# finding verb frequency distribution
verb_fd = nltk.FreqDist(word for (word,tag) in tags if tag == 'VBG')
verb_fd
```

```python
# Reading top Ten most common words in verb frequency distribution 
stemmer_input = [ word for( word, value ) in verb_fd.most_common(10) if word ]
```

```python
stemmer_output = [snowball.stem(word) for word in stemmer_input]
df =  pd.DataFrame(list(zip(stemmer_input,stemmer_output)))
df
```
